```python
#Используется для записи модели в файл
from sklearn.externals import joblib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def get_correct_input(data):
    gender=pd.get_dummies(data['Sex'])
    embarked=pd.get_dummies(data['Embarked'], prefix='Embarked')
    pcl=pd.get_dummies(data['Pclass'], prefix='Pclass')
    age=pd.get_dummies(pd.cut(data.Age, 10, labels=False).astype(int), prefix='Age')
    fare=pd.get_dummies(pd.cut(data.Fare, 10, labels=False).astype(int), prefix='Fare')
    return pd.concat([gender,embarked,pcl,age,fare],axis=1)
```

```python
gender=pd.get_dummies(data['Sex'])
embarked=pd.get_dummies(data['Embarked'], prefix='Embarked')
pcl=pd.get_dummies(data['Pclass'], prefix='Pclass')
```

```python
pcl.head()
```

```python
age=pd.get_dummies(pd.cut(data.Age, 10, labels=False), prefix='Age')
fare=pd.get_dummies(pd.cut(data.Fare, 10, labels=False), prefix='Fare')
```

```python
data.head()
```

```python
data_inputs=pd.concat([gender,embarked,pcl,age,fare],axis=1)
data_inputs.head()
```

```python
get_correct_input(data).head()
```

```python
expected_output = data[["Survived"]]
expected_output.head()
```

```python
from sklearn import svm
clf = svm.SVC(gamma=0.001)
```

```python
clf.fit(data_inputs, expected_output.values.ravel())
```

```python
joblib.dump(clf, "titanic_model1", compress=9)
```
